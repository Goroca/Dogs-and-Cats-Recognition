```python
import os
import matplotlib.pyplot as plt
from matplotlib.image import imread
from PIL import Image # For handling the images
```

```python
# Define the directory to be inspected as a list
list_train = os.listdir("train")

# Obtain total number of images
number_files_train = len(list_train)
print('Number of images: ' + str(number_files_train))

number_cats_train = 0
number_dogs_train = 0

# Iterate directory and print and count images of each class
for image in list_train:
    
    if image[:3] == 'cat':
        
        number_cats_train = number_cats_train + 1
        
    else:
        
        number_dogs_train = number_dogs_train + 1   
        
    print(image)     
    
    
print('Number of cat images: ' + str(number_cats_train))
print('Number of dog images: ' + str(number_dogs_train))    
```

```python
# Define the directory to be inspected as a list
list_test = os.listdir('test1')

#Obtain total number of images
number_files_test = len(list_test)
print('Number of images: ' + str(number_files_test))

# Iterate directory 
for image in list_test:   
        
    print(image)     
    
```

```python
# Plot dog photos from the dogs vs cats dataset

# Define location of dataset
folder = 'train/'

# Plot first few images
for i in range(9):
    # Define subplot
    plt.subplot(330 + 1 + i)
    # Define filename
    filename = folder + 'dog.' + str(i) + '.jpg'
    # Load image pixels
    image = imread(filename)
    # Plot raw pixel data
    plt.imshow(image)
    
# Show the figure
plt.show()
```

```python
# Plot cat photos from the dogs vs cats dataset

# Define location of dataset
folder = 'train/'
# Plot first few images
for i in range(9):
    # Define subplot
    plt.subplot(330 + 1 + i)
    # Define filename
    filename = folder + 'cat.' + str(i) + '.jpg'
    # Load image pixels
    image = imread(filename)
    # Plot raw pixel data
    plt.imshow(image)
# Show the figure
plt.show()
```

```python
folder = 'train'

count=0

for f_name in os.listdir(folder):
    if(count<20):
        img = Image.open(folder + '/' + f_name)
        plt.imshow(img, vmin=0, vmax=255)
        plt.title('subject: ' + str(f_name))
        plt.show()
        count+=1
```
